```python
import pandas as pd
data = pd.read_csv("/content/breastcancer.csv")

data.head(6)
```

```python
data.describe()
```

# **Convert categorical data into numerical data**

```python
from sklearn.preprocessing import OneHotEncoder

# creating a object of onehotencoder
encoder = OneHotEncoder()

data['diagnosis'] = data['diagnosis'].astype( 'category' )
data['diagnosis'] = data['diagnosis'].cat.codes

data.head(30)
```

# **Splitting Dataset into training and testing Dataset**

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X = data.drop(['id', 'diagnosis', 'target'], axis=1)
Y = data['diagnosis']

x_train, x_test, y_train, y_test = train_test_split(X, Y,test_size = 0.25, random_state = 2)
```

# **Standardization on the Training and Testing data**

```python
from sklearn.preprocessing import StandardScaler
# Standardization on training and testing data
sc = StandardScaler()
X_train = sc.fit_transform(x_train)
X_test = sc.fit_transform(x_test)
```

# **Perform Decision Tree on the given dataset**

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier

dtree = DecisionTreeClassifier()
dtree.fit(X_train, y_train)
```

```python
y_pred = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

# Confusion Matrix On the prediction and actual data
cm = confusion_matrix(y_test, y_pred)
print("Confusion matrix:\n",cm,"\n\n")

# Classification report on the prediction
creport = classification_report(y_test, y_pred)
print("Classification report:\n",creport)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=dtree.classes_)
disp.plot()
plt.show()
```

# **Perform Random Forest on the dataset**

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(max_depth=2, random_state=0)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

# Confusion Matrix On the prediction and actual data
cm = confusion_matrix(y_test, y_pred)
print("Confusion matrix:\n",cm,"\n\n")

# Classification report on the prediction
creport = classification_report(y_test, y_pred)
print("Classification report:\n",creport)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
import matplotlib.pyplot as plt
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=dtree.classes_)
disp.plot()
plt.show()
```
