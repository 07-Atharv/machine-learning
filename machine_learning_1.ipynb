# Spliting the data into training and testing

```python
from sklearn.model_selection import train_test_split
#To divide the data into train and test we have used these library
store_x = data_store.drop(['id','diagnosis','target'],axis=1)
store_y = data_store['target']
x_train , x_test , y_train , y_test = train_test_split(store_x,store_y,test_size=0.20,random_state=2)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```

**Using select k best method for feature selection**

```python
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest
#k is the number of features which we want as high valuable features
store_k_best_feature = SelectKBest(score_func=chi2,k=5)
#used to train the data
store_k_best_feature.fit(x_train,y_train)
#columns is used to get the features which is returned by get support in the form of integer index
features = x_train.columns[store_k_best_feature.get_support()]
print(features)
```

**Using select percentile method for feature selection**

```python
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectPercentile
store_best = SelectPercentile(score_func=chi2,percentile=50)
store_best.fit(x_train,y_train)
features=x_train.columns[store_best.get_support()]
print(features)
#The same concept is used here of feature selection but the method is different
```
